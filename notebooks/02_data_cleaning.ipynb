```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df["TotalCharges"]=pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df=df.dropna()
```

```python
df=df.drop_duplicates()
```

```python
df.drop(columns="customerID", inplace=True)
```

```python
df.to_csv("../data/processed/customer_churn_clean.csv", index=False)
```

# Data Cleaning Summary

- Converted `TotalCharges` from object to float.
- Replaced blank strings with NaN using `pd.to_numeric(errors="coerce")`.
- Removed 11 rows with missing `TotalCharges`.
- Verified there were no duplicate records.
- Removed the `customerID` column because it is only an identifier.
- Saved the cleaned dataset for further analysis.
